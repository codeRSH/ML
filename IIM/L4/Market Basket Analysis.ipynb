```python
#Importing libraries

import pandas as pd
import numpy as np
from apyori import apriori
```

```python
#Read clean data
df = pd.read_csv('C:\\Users\\ravs\\Development\\ML\\IIM\L4\\Market_Basket_Optimisation.csv',header=None)
```

```python
df.head()
```

```python
#Replacing NaN with 0
df.fillna(0,inplace=True)
```

```python
df.head()
```

```python
#Creating a list 
#i is for outer loop and j is for inner loop
#i move vertically and j move horizontally 
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
transactions[5]
```

```python
rules = apriori(transactions,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
```

```python
print(rules)
```

```python
Results = list(rules)
```

```python
df_results = pd.DataFrame(Results)
```

```python
df_results.head()
```

```python
support = df_results.support
```

```python
support
```

```python
first_values = []
second_values = []
third_values = []
fourth_value = []
for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
```

```python
lhs = pd.DataFrame(first_values)
rhs= pd.DataFrame(second_values)
confidance=pd.DataFrame(third_values,columns=['Confidance'])
lift=pd.DataFrame(fourth_value,columns=['lift'])
```

```python
rhs
```

```python
df_final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
```

```python
df_final
```

```python
df_final.fillna(value=' ', inplace=True)
```

```python
df_final.columns = ['lhs',1,'rhs','rhs_1','rhs_2','support','confidance','lift']
```

```python
df_final
```

```python
df_final['lhs'] = df_final['lhs']+str(", ")+df_final[1]
```

```python
df_final.drop(columns=[1],inplace=True)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df1 = pd.read_csv('C:\\RecogX_Backup\\Deepak\\IIM\\Eruditas\\data.csv', encoding="ISO-8859-1")
df1.head()
```

```python
df1.Country.value_counts().head(5)
```

```python
df1 = df1[df1.Country == 'France']
```

```python
df1['Description'] = df1['Description'].str.strip()
```

```python
df1.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df1['InvoiceNo'] = df1['InvoiceNo'].astype('str')
df1 = df1[~df1['InvoiceNo'].str.contains('C')]
```

```python
df1 = df1[df1.Quantity >0]
```

```python
basket = (df1[df1['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket
```

```python
basket['ALARM CLOCK BAKELIKE RED'].head(10)
```

```python
def convert_into_binary(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
basket_sets = basket.applymap(convert_into_binary)
```

```python
basket_sets['ALARM CLOCK BAKELIKE RED'].head()
```

```python
basket_sets['POSTAGE']
```

```python
 basket_sets.drop(columns=['POSTAGE'],inplace=True)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules_mlxtend
```

```python
rules_mlxtend[ (rules_mlxtend['lift'] >= 6) &
       (rules_mlxtend['confidence'] >= 0.8) ]
```
