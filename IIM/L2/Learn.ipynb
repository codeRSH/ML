```python

# Import Packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python

# Read the Pizza1 file
data = pd.read_csv("pizza1.csv")
print(data)
```

```python

data.isnull().sum()
```

```python
data.shape
```

```python
# Drop the rows with null value

data1 = data.dropna()
data1.shape
```

```python
# Replace NAN values by their respective column mean

data2 = data.fillna(data.mean())
data2.isnull().sum()
```

```python
data.head()
```

```python
data2.head()
```

```python
# Replace NAN values by their respective column median

data3 = data.fillna(data.median())
data3.head()
```

```python
data4 = data.fillna(0)
data4.head()
```

```python
# Dealing with Column Reduction

# Dropping correlated columns

corr_matrix = data3.corr()
corr_matrix
```

```python
# Visualizing the correlation matrix

sb.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k = 1).astype(np.bool))
upper
```

```python
# Find columns which have correlation of greater than 0.7
drp = [column for column in upper.columns if any(upper[column] > 0.70)]
drp
```

```python
data5 = data3.drop(data[drp], axis = 1)

data5
```

```python
# Implement PCA for Dimension Reduction

# Reading the datasets

dataset = pd.read_csv("Wine.csv")
```

```python
dataset.head(5)
```

```python
# Create a dataframe of Independent and Dependent variables. 

x = dataset.loc[:, 'Alcohol':'Proline']
x
```

```python
y = dataset.loc[:, 'Customer_Segment']
y
```

```python
# Preprocess the dataset standardize the dataset

from sklearn import preprocessing

sc = preprocessing.StandardScaler().fit(x)
sc
```

```python
sc.mean_
```

```python
sc.scale_
```

```python
x = sc.transform(x)
x
```

```python
from sklearn.decomposition import PCA

pca = PCA()

results = pca.fit(x)
```

```python
Z = results.transform(x)
Z
```

```python
plt.plot(results.explained_variance_ratio_ * 100)
```
